```python
import os
import pandas as pd
import numpy as np
from sklearn.metrics import f1_score, confusion_matrix
from matplotlib import pyplot as plt
import re
import seaborn as sbs
```

# Setting Up Filter

```python
def keyword_filter(keyword, dataset):
    
    keyword_labels = list()
    
    cols = dataset.columns
    
    if 'text' in cols:
        column = dataset.text
        
    elif 'subtitle' in cols:
        column = dataset.subtitle
        
    elif 'title' in cols:
        column = dataset.title
    
    else:
        raise ValueError("No valid text columns.")
    
    
    for t in column:
        match = re.findall(pattern, str(t), re.IGNORECASE)
        keyword_labels.append((len(match) > 0)*1)
        
    return keyword_labels
```

```python
pattern = r'gender'
```

# Accuracy Metrics from Labeled Datsets for Keyword Filter

```python
train_path = ('/home/pranavgoel/trans-fer-entropy/internal_relevance_annotation/' + 
              'train_sample_for_relevance_classifier.csv')
test_path =  ('/home/pranavgoel/trans-fer-entropy/internal_relevance_annotation/' +
              'test_sample_for_relevance_classifier.csv')
```

```python
train_df = pd.read_csv(train_path)
test_df = pd.read_csv(test_path)
```

```python
sbs.heatmap(confusion_matrix(train_df.label, keyword_filter(pattern, train_df)), annot=True, cmap='gray_r')
plt.title('Train Dataset')
```

```python
sbs.heatmap(confusion_matrix(test_df.label, keyword_filter(pattern, test_df)), annot=True, cmap='gray_r')
plt.title('Test Dataset')
```

# Filtering All Datasets

```python
indir = "/home/pranavgoel/trans-fer-entropy/obtaining_news_collections/data/"
files = os.listdir(indir)
files
```

```python
outdir= "/data_users1/sagar/trans-fer-entropy/kw_filtered_data/"
```

```python
pattern = r'gender'

for f in files:
    
    if f[-4:] == '.csv':
        
        print(f"Processing {f}")
        
        df = pd.read_csv(indir+f)
        
        labels = keyword_filter(pattern, df)
        
        df['gender_label'] = labels
        
        df_new = df[df.gender_label == 1]
        
        df_new.to_csv(outdir+f, sep=',')
```
