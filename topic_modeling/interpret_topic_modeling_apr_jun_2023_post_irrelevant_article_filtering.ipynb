```python
#note - this is working with the files post the fox-news-fix
import little_mallet_wrapper as lmw
import pandas as pd

from IPython.display import clear_output
import pickle
```

```python
MALLET_OUTPUT_PATH = '/home/pranavgoel/mallet_output/trans_fer_entropy_apr_jun_2023_post_irrelevant_article_filtering/'
```

```python
combined_df = pd.read_csv('apr_jun_2023_post_irrelevant_article_filtering/all_texts_combined_post_irrelevant_article_filtering.csv')
print(combined_df.info())
```

```python
titles_and_texts = list(zip(combined_df['title'], 
                            combined_df['text']))

titles_and_texts = list(map(lambda x:x[0] + '\n\n\n' + x[1], 
                            titles_and_texts))
```

```python
def display_top_words_and_top_docs_for_specific_topic_ind(titles_and_texts, 
                                                          topic_keys, 
                                                          topic_distributions, 
                                                          topic_ind, 
                                                          num_top_words = 25,
                                                          num_top_docs = 5):
    print('Topic ' + str(topic_ind))
    print('\nTop words:\n' + ', '.join(topic_keys[topic_ind][:num_top_words]))
    print('\n\nTop documents: ')
    top_doc_ind = 0
    for _, d in lmw.get_top_docs(titles_and_texts, topic_distributions, topic_index=topic_ind, n=num_top_docs):
        print(str(top_doc_ind+1) + '. ' + d[:2000] + '.........')
        print('\n=== ===\n')
        top_doc_ind += 1
```

```python
num_topics = 20
topic_keys = lmw.load_topic_keys(MALLET_OUTPUT_PATH + '/mallet.topic_keys.' + str(num_topics))
topic_distributions = lmw.load_topic_distributions(MALLET_OUTPUT_PATH + '/mallet.topic_distributions.' + str(num_topics))
```

```python
#displaying earlier topic labels wo fox news fix, expecting many labels might get reused
prev_topic_labels = pickle.load(open('../topic_labels_20_finalized.pkl', 'rb'))
for t in prev_topic_labels:
    print(str(t) + ': ' + prev_topic_labels[t].replace('\n', ''))
```

```python
proposed_labels = []
```

```python
#forced to rerun this as jupyter notebook malfunctions and sometimes does not ask for input 
for topic_ind in range(len(proposed_labels), num_topics):
    clear_output(wait=True)
    display_top_words_and_top_docs_for_specific_topic_ind(titles_and_texts,
                                                          topic_keys, 
                                                          topic_distributions,
                                                          topic_ind)
    print('\n---')
    proposed_label = str(input('Enter label: '))
    
    proposed_labels.append(proposed_label)
```

```python
#create_and_save_top_words_and_top_docs_and_proposed_labels
topic_inds, top_words, top_docs = [], [], []
for topic_ind in range(num_topics):
    topic_inds.append(topic_ind)
    top_words.append(', '.join(topic_keys[topic_ind][:20]))
    top_doc_texts = []
    top_doc_ind = 0
    for _, d in lmw.get_top_docs(titles_and_texts, topic_distributions, topic_index=topic_ind, n = 5):
        top_doc_texts.append(str(top_doc_ind+1) + '. ' + d[:2000] + '.........')
        top_doc_ind += 1
    top_docs.append('\n=== ===\n'.join(top_doc_texts))
df = pd.DataFrame()
df['Topic Index'] = topic_inds
df['Top Words'] = top_words
df['Top Docs'] = top_docs
df['Proposed Label (First Draft)'] = proposed_labels
print(df.info())
```
