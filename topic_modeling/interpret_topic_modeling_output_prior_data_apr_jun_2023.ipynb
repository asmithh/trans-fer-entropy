### This notebook looks at the topics (in terms of top words and top documents) for each topic, to enable interpretation for each topic uncovered by various topic models trained and saved in train_mallet_on_combined_article_text_data_apr_jun_2023_data.py

```python
import little_mallet_wrapper as lmw
import pandas as pd
```

```python
MALLET_OUTPUT_PATH = '/home/pranavgoel/mallet_output/trans_fer_entropy_apr_jun_2023/'
```

```python
combined_df = pd.read_csv('prior_data_apr_jun_2023/all_texts_combined.csv')
print(combined_df.info())
```

```python
titles_and_texts = list(zip(combined_df['title'], 
                            combined_df['text']))

titles_and_texts = list(map(lambda x:x[0] + '\n\n\n' + x[1], 
                            titles_and_texts))
```

```python
def display_top_words_and_top_docs_for_specific_topic_ind(titles_and_texts, 
                                                          topic_keys, 
                                                          topic_distributions, 
                                                          topic_ind, 
                                                          num_top_words = 20,
                                                          num_top_docs = 5):
    print('Topic ' + str(topic_ind))
    print('\nTop words:\n' + ', '.join(topic_keys[topic_ind][:num_top_words]))
    print('\n\nTop documents: ')
    top_doc_ind = 0
    for _, d in lmw.get_top_docs(titles_and_texts, topic_distributions, topic_index=topic_ind, n=num_top_docs):
        print(str(top_doc_ind+1) + '. ' + d[:2000] + '.........')
        print('\n=== ===\n')
        top_doc_ind += 1
```
