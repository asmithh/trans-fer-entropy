```python
import numpy as np
plt.rcParams["figure.figsize"] = (10,30)
plt.rcParams.update({'font.size': 12})
import matplotlib.pyplot as plt
import pickle
z = pickle.load(open('../../trans-fer-entropy/transfer/filtered/filtered_medicloud_ete_results_20_pubs_on_states.pkl', 'rb'))

W = 2
plt.tight_layout()
fig, ax = plt.subplots(10, 2)

VMIN = -1
VMAX = 1
states = ['california', 'illinois', 'newyork', 'ohio', 'florida',  'texas']
pubs = ['foxnews.com', 'nytimes.com']
for i in range(20):
    mtx = np.zeros((6, 2))
    for x, s in enumerate(states):
        for y, p in enumerate(pubs):
            mtx[x, y] = z[i][(p, s, W, 'x_on_y')] - z[i][(p, s, W, 'shuffled_x_on_y')]
    ax[int(np.floor(i / 2.0)), int(i % 2.0)].set_title(topic_labels_20[i])
    if i == 19:
        print(np.mean(mtx, axis=0))
    # ax[int(np.floor(i / 2.0)), int(i % 2.0)].set_xticks([0, 1], ['FOX', 'NYT'], rotation=40)
    ax[int(np.floor(i / 2.0)), int(i % 2.0)].set_yticks([i for i in range(2)], pubs)
    ax[int(np.floor(i / 2.0)), int(i % 2.0)].set_xticks([i for i in range(6)], states, rotation=90)
    poo = ax[int(np.floor(i / 2.0)), int(i % 2.0)].imshow(mtx.transpose(), vmin=VMIN, vmax=VMAX, cmap='viridis')
fig.colorbar(poo)
plt.tight_layout()
plt.savefig("filtered_20_pubs_on_states_ete.svg")
plt.savefig("filtered_20_pubs_on_states_ete.png")
```
