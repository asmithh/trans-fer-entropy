```python
import numpy as np
plt.rcParams["figure.figsize"] = (10,30)
plt.rcParams.update({'font.size': 12})

W = 2
plt.tight_layout()
fig, ax = plt.subplots(10, 2)

VMIN = -1
VMAX = 1
states = ['california', 'florida', 'illinois', 'newyork', 'ohio', 'texas']
pubs = ['foxnews.com', 'nytimes.com']
for i in range(20):
    mtx = np.zeros((6, 2))
    for x, s in enumerate(states):
        for y, p in enumerate(pubs):
            mtx[x, y] = z[i][(s, p, W, 'y_on_x')] - z[i][(s, p, W, 'shuffled_y_on_x')]
    ax[int(np.floor(i / 2.0)), int(i % 2.0)].set_title(topic_labels_20[i])
    if i == 19:
        print(np.mean(mtx, axis=0))
    ax[int(np.floor(i / 2.0)), int(i % 2.0)].set_xticks([0, 1], ['FOX', 'NYT'], rotation=40)
    ax[int(np.floor(i / 2.0)), int(i % 2.0)].set_yticks([i for i in range(6)], states)
    poo = ax[int(np.floor(i / 2.0)), int(i % 2.0)].imshow(mtx, vmin=VMIN, vmax=VMAX, cmap='coolwarm')
fig.colorbar(poo)
plt.tight_layout()
plt.savefig("20_pubs_on_states_ete.svg")
plt.savefig("20_pubs_on_states_ete.png")
```
